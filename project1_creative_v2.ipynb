```python
from sklearn.preprocessing import MinMaxScaler
import umap
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from tqdm import tqdm
```

```python
# read the data created by Project1_data.ipynb

train = pd.read_csv('train_vectors.csv')
validation = pd.read_csv('validation_vectors.csv')
```

```python
# drop columns name and transform labels

new_train = train[[col for col in train.columns if col not in ['name', 'label']]]
labels_dict = {'cultural representative': 0, 'cultural exclusive': 1, 'cultural agnostic': 2}
new_y = train.replace({'label': labels_dict})
new_y = new_y['label']

new_validation = validation[[col for col in validation.columns if col not in ['name', 'label']]]
new_validation_y = validation.replace({'label': labels_dict})
new_validation_y = new_validation_y['label']
```

```python
# weight vector
weight_vec = [1,1]
weight_vec += [1/19]*19
weight_vec += [1/112]*112
weight_vec += [1/100]*100
```

#### Grid search

```python
pipeline = Pipeline([
    ('umap', umap.UMAP(random_state=42, metric='euclidean', learning_rate=0.01)),
    ('knn', KNeighborsClassifier())
])

# testing over usual values of the parameters
param_grid = {
    'umap__n_neighbors': [10, 20, 30, 40, 50],
    'umap__n_components': [2, 3, 5, 10, 15, 20, 30],
    'knn__n_neighbors': list(range(5,50,5))
}
```

```python
grid = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy', verbose=2, n_jobs=-1)
tqdm(grid.fit(new_train, new_y))
```

```python
pd.DataFrame(grid.cv_results_).to_csv('grid_search_results.csv', index=False)
```

```python
print("Best parameters found:", grid.best_params_)
```

### Testing best parameters

```python
# umap test with the best parameters
reducer = umap.UMAP(metric='euclidean', random_state=42, n_neighbors=50, learning_rate=0.01, n_components=5)
embedding = reducer.fit_transform(new_train)
```

```python
# validation projection

validation_embedding = reducer.transform(new_validation)
```

```python
# knn test with the best parameters
knn = KNeighborsClassifier(n_neighbors=45)
knn.fit(embedding, new_y)
y_pred = knn.predict(validation_embedding)

# evaluation
accuracy = accuracy_score(new_validation_y, y_pred)
f1 = f1_score(new_validation_y, y_pred, average='weighted')
precision = precision_score(new_validation_y, y_pred, average='weighted')
recall = recall_score(new_validation_y, y_pred, average='weighted')


print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'F1 Score: {f1 * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
print(f'Recall: {recall * 100:.2f}%')
```

### Scaled and weighted grid search

```python
pipeline2 = Pipeline([
    ('umap', umap.UMAP(random_state=42, metric='euclidean', learning_rate=0.01)),
    ('knn', KNeighborsClassifier())
])

param_grid = {
    'umap__n_neighbors': [10, 20, 30, 40, 50],
    'umap__n_components': [2, 3, 5, 10, 15, 20, 30],
    'knn__n_neighbors': list(range(5,50,5))
}
```

```python
# scaler for the new training set
scaler = MinMaxScaler(feature_range=(0, 1))

new_train = pd.DataFrame(scaler.fit_transform(new_train), columns=new_train.columns)

# apply the weight vector to the new training set
new_train = new_train * np.sqrt(weight_vec)
```

```python
grid2 = GridSearchCV(pipeline2, param_grid, cv=10, scoring='accuracy', verbose=2, n_jobs=-1)
tqdm(grid2.fit(new_train, new_y))
```

```python
pd.DataFrame(grid2.cv_results_).to_csv('grid2_search_results.csv', index=False)
```

```python
print("Best parameters found:", grid2.best_params_)
```

### Testing best parameters of the second grid

```python
# umap test with the best parameters
reducer2 = umap.UMAP(metric='euclidean', random_state=42, n_neighbors=40, learning_rate=0.01, n_components=5)
embedding = reducer2.fit_transform(new_train)
```

```python
# validation projection

validation_embedding = reducer2.transform(new_validation)
```

```python
# knn test with the best parameters
knn2 = KNeighborsClassifier(n_neighbors=25)
knn2.fit(embedding, new_y)
y_pred = knn2.predict(validation_embedding)

# evaluation
accuracy = accuracy_score(new_validation_y, y_pred)
f1 = f1_score(new_validation_y, y_pred, average='weighted')
precision = precision_score(new_validation_y, y_pred, average='weighted')
recall = recall_score(new_validation_y, y_pred, average='weighted')


print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'F1 Score: {f1 * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
print(f'Recall: {recall * 100:.2f}%')
```
